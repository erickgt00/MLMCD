## Mediana de la Tasa de Interes

```python
v2 = df["Interest Rate"]
vector2 = v2.values
vector2.sort()
n = len(vector2) # Calcula el tamaño del vector
if n % 2 == 0:
    mediana = (vector2[n // 2] + vector2[n // 2 - 1]) / 2
else:
    mediana = vector2[n // 2]

print(mediana)
```

## La moda del Titulo del Prestamo

```python
v3 = df["Loan Title"]
vector3 = v3.values
# Abrimos una librería vacía
frec = {}

# Recorremos el conjunto y contamos la frecuencia
for j in vector3:
    if j in frec:
        frec[j] += 1
    else:
        frec[j] = 1

moda = max(frec, key=frec.get)

print(moda)
```

## Prueba de Hipotesis Kolmogorov-Smirnov (K-S)

La prueba K-S compara la distribución empírica (la distribución de los datos) con una distribución teórica especificada (como una distribución normal). El valor p de la prueba indica la probabilidad de observar una diferencia tan grande o mayor entre las dos distribuciones si las dos distribuciones son realmente iguales. Esta es utilizada dado a que nuestros datos tienen mas de 1,000 observaciones.

### Prueba de Normalidad para el Monto Prestado

```python
# Creamos una funcion para rechazar o no nuestra hipotesis nula
def rechazo(p, alpha):
    if p < alpha:
        print("Se rechaza la hipótesis nula.")
    else:
        print("No se rechaza la hipótesis nula.")
```

```python
from scipy.stats import kstest

data = df["Loan Amount"]
statistic, p = kstest(data, 'norm')
alpha = 0.05

rechazo(p, alpha)
```

Por lo tanto podemos concluir que nuestros datos provienen de alguna otra distribución.

```python
import matplotlib.pyplot as plt

plt.hist(data)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma del Monto Prestado')
plt.show()
```

### Prueba para el Total de Intereses Recividos

En esta haremos una prueba logaritmica

```python
from scipy.stats import expon, kstest
data2 =df["Total Received Interest"] 

# ajustar una distribución log-normal a los datos
param = expon.fit(data2)

statistic, p = kstest(data2, 'expon', args=param)

alpha = 0.05

rechazo(p, alpha)
```

Por lo tanto podemos concluir que nuestros datos provienen de alguna otra distribución.

```python
plt.hist(data2)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Total de Interéses Recividos')
plt.show()
```
